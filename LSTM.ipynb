```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load the data from the CSV file
data = pd.read_csv('Project.csv')

# Specify the feature names and output variable
feature_names = ['CUSTOMER_ID', 'READING_DATETIME', 'CALENDAR_KEY', 'GENERAL_SUPPLY_KWH', 'Day_Month',
                 'DayOfWeek', 'Weekday_or_Weekend', 'Time interval - 30mins', 'Time interval -1hr',
                 'Seasons', 'Holiday']
output_variable = 'GENERAL_SUPPLY_KWH'

# Extract the input features and output variable
X = data[feature_names]
y = data[output_variable]

# Handle non-numeric data
encoder = LabelEncoder()
X_encoded = X.copy()
for feature in feature_names:
    if X[feature].dtype == 'object':
        X_encoded[feature] = encoder.fit_transform(X[feature])

# Scale the input features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Reshape the data to match the LSTM input shape
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Build and compile the LSTM model
model = Sequential()
model.add(LSTM(128, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True))
model.add(LSTM(64))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=64)

# Evaluate the model on the test set
loss = model.evaluate(X_test, y_test)

# Make predictions
predictions = model.predict(X_test)

# Calculate R-squared
r2 = r2_score(y_test, predictions)

# Calculate MAE
mae = mean_absolute_error(y_test, predictions)

print(f"Test Loss: {loss}")
print(f"R-squared: {r2}")
print(f"MAE: {mae}")
```
